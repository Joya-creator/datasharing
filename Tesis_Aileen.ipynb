```python
# prompt: CORELACION ENTRE CATARATA Y SINDROME METABOLICO

# Assuming 'SM' column represents Sindrome Metabolico (Metabolic Syndrome)
# and it's binary (1 for presence, 0 for absence)

# Calculate the correlation
correlation = df['Catarata'].corr(df['Sx Metabolico'])

print("Correlacion entre Cirugía de Catarata and Sindrome Metabolico:", correlation)

contingency_table = pd.crosstab(df['Catarata'], df['Sx Metabolico'])
print(contingency_table)

sns.heatmap(contingency_table, annot=True, cmap="YlGnBu")
plt.show()
```

```python
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-squared statistic:", chi2)
print("P-value:", p)
```
